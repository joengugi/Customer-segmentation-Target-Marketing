```python
## Data Exploration and Preprocessing
import pandas as pd

    # Loading the dataset
data = pd.read_excel("marketing_campaign.xlsx")

    # Display the first few rows and dataset information
data.head(20)
```

```python
data.info()
```

```python
data.describe()
```

```python
## Data Preprocessing

# Checking the percentage of missing values
missing_values = data.isnull().sum() / len(data) * 100
print(missing_values)
```

```python
# Dropping columns with a high percentage of missing values

# Columns with at least 30% missing values will be dropped
threshold = 30
columns_to_drop = missing_values[missing_values > threshold].index
data_cleaned = data.drop(columns=columns_to_drop, axis=1)
data_cleaned.info()
```

```python
# Filling missing values for numerical columns with median
numerical_cols = data_cleaned.select_dtypes(include=['float64', 'int64']).columns
data_cleaned[numerical_cols] = data_cleaned[numerical_cols].fillna(data_cleaned[numerical_cols].median())

# Encoding categorical variables with one-hot encoding (if any categorical variables are present)
categorical_cols = data_cleaned.select_dtypes(include=['object']).columns
data_encoded = pd.get_dummies(data_cleaned, columns=categorical_cols, drop_first=True)
data_encoded.info()
```

```python
# Scaling numerical variables for uniformity
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_encoded[numerical_cols] = scaler.fit_transform(data_encoded[numerical_cols])

# Verifying the cleaned and processed data
data_encoded.info()
```

```python
## Task Selection
## Task selected: Classification
## Variable to predict: Response
 
## Classifier 1: Random Forest

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Checking the unique values and data type of the 'Response' column
#3 y_unique_values = y.unique()
## y_dtype = y.dtype

# Splitting data into features (X) and target (y)
X = data_encoded.drop('Response', axis=1)
y = data_encoded['Response']

# Ensuring the 'Response' column is categorical
y = y.astype('int')

# Splitting the dataset into training and testing sets
model = RandomForestClassifier(random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
model.fit(X_train, y_train)

# Making predictions
y_pred = model.predict(X_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Classification Evaluation
print("Classification Report")
print(classification_report(y_test, y_pred))

print("Confusion Matrix")
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Reds')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

## Classifier 2: Logistic Regression
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train, y_train)
y_pred_logistic = logistic_model.predict(X_test)

# Metrics for Logistic Regression
print("Logistic Regression:\n")
print(classification_report(y_test, y_pred_logistic))
print("Accuracy:", accuracy_score(y_test, y_pred_logistic))

# Confusion Matrix Visualization for Logistic Regression
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred_logistic), annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

## Classifier 3: K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors = int(np.sqrt(len(X_train))))
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)

# Metrics for KNN
print("K-Nearest Neighbors (KNN):\n")
print(classification_report(y_test, y_pred_knn))
print("Accuracy:", accuracy_score(y_test, y_pred_knn))

# Confusion Matrix Visualization for KNN
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred_knn), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - KNN')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.cluster import KMeans

# TASK 2: Clustering
# Selecting relevant features for clustering (e.g., spending patterns)
clustering_features = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
clustering_data = data_encoded[clustering_features]
```

```python
# Elbow Method to determine optimal clusters
inertia = []
k_values = range(1, 8)

for k in k_values:
    kmeans_temp = KMeans(n_clusters=k, random_state=42)
    kmeans_temp.fit(clustering_data)
    inertia.append(kmeans_temp.inertia_)

# Plotting the Elbow Curve
plt.figure(figsize=(10, 8))
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(clustering_data)

# Adding cluster labels to the original data
data_encoded['Cluster'] = clusters

# Visualizing Clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x=data_encoded['MntWines'], 
    y=data_encoded['MntMeatProducts'], 
    hue=data_encoded['Cluster'], 
    palette='Set2')
plt.title('Customer Clusters Based on Spending Patterns')
plt.xlabel('MntWines')
plt.ylabel('MntMeatProducts')
plt.legend(title='Cluster')
plt.show()
```

```python
# Analyzing Cluster Characteristics
print("Clustering Analysis")
print(data_encoded.groupby('Cluster')[clustering_features].mean())
```

```python
# Heatmap of cluster centers
cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns=clustering_features)
plt.figure(figsize=(8, 6))
sns.heatmap(cluster_centers, annot=True, cmap='hot', fmt='.1f')
plt.title('Cluster Centers (Spending Patterns)')
plt.xlabel('Features')
plt.ylabel('Clusters')
plt.show()
```
